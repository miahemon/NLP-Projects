```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = text.lower()
    tokens = word_tokenize(text)
    stopwords_set = set(stopwords.words('english'))
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stopwords_set]
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text
```

```python
preprocess_data=lambda text:preprocess_text(text)
train_df["text_prepro"]=train_df["Text"].apply(preprocess_data)
```

**Perform:**

```python
embedding_matrix = np.zeros((len(word2vec_model.wv.key_to_index) + 1, word2vec_model.vector_size))
for word, i in word2vec_model.wv.key_to_index.items():
    embedding_vector = word2vec_model.wv[word]
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, Dense, MultiHeadAttention, GlobalMaxPooling1D, LayerNormalization, Dropout
from tensorflow.keras.initializers import GlorotUniform
from tensorflow.keras.regularizers import l2

def create_model(max_length, vocab_size, embedding_dim, num_heads, ff_dim, dropout_rate, batch_size, num_classes, learning_rate):
    inputs = Input(shape=(max_length,))
    embedding = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(inputs)
    x = Dropout(dropout_rate)(embedding)
    multi_head_attention = MultiHeadAttention(num_heads=num_heads, key_dim=embedding_dim)
    x = multi_head_attention(query=x, value=x, key=x)
    x = LayerNormalization()(x)
    ff_network = Dense(ff_dim, activation='relu', kernel_initializer=GlorotUniform(), kernel_regularizer=l2(0.01))(x) 
    ff_network = Dense(embedding_dim, kernel_initializer=GlorotUniform(), kernel_regularizer=l2(0.01))(ff_network)
    x = x + ff_network
    x = LayerNormalization()(x)
    x = GlobalMaxPooling1D()(x)
    x = Dropout(0.1)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
from scipy.stats import randint, uniform
param_dist = {
    'num_heads': [4, 8, 16, 32],
    'ff_dim': randint(64, 512),
    'dropout_rate': [0.1, 0.2, 0.3, 0.4],
    'batch_size': [32, 64, 128, 256],
    'learning_rate': uniform(0.0001, 0.01)  # Learning rate range from 0.0001 to 0.01
}
```

```python
from sklearn.metrics import accuracy_score

max_length = max_length
vocab_size = embedding_matrix.shape[0]
embedding_dim = embedding_matrix.shape[1]
num_classes = jumlah_kelas

class CustomEstimator:
    def __init__(self, num_heads=8, ff_dim=128, dropout_rate=0.2, batch_size=32, learning_rate=0.001):
        self.num_heads = num_heads
        self.ff_dim = ff_dim
        self.dropout_rate = dropout_rate
        self.batch_size = batch_size
        self.learning_rate = learning_rate
        self.model = None

    def fit(self, X, y, **kwargs):
        if self.model is None:
            self.model = self._init_model()
        if self.model is not None:
            self.model.fit(X, y, batch_size=self.batch_size, **kwargs)
        else:
            print("Model is not initialized. Skipping fitting.")

    def predict(self, X):
        if self.model is not None:
            y_pred_prob = self.model.predict(X)
            y_pred = np.argmax(y_pred_prob, axis=1)
            return y_pred
        else:
            print("No model to make predictions. Continuing to the next step of the algorithm.")
            return None

    def score(self, X, y):
        if self.model is not None:
            y_pred = self.predict(X)
            return accuracy_score(y, y_pred)
        else:
            print("No model to compute score. Continuing to the next step of the algorithm.")
            return 0.0

    def _init_model(self):
        return create_model(max_length, vocab_size, embedding_dim, self.num_heads, self.ff_dim, self.dropout_rate, self.batch_size, num_classes, self.learning_rate)

    def get_params(self, deep=True):
        return {
            'num_heads': self.num_heads,
            'ff_dim': self.ff_dim,
            'dropout_rate': self.dropout_rate,
            'batch_size': self.batch_size,
            'learning_rate': self.learning_rate
        }

    def set_params(self, **params):
        for param, value in params.items():
            setattr(self, param, value)
        return self
```

```python
from sklearn.model_selection import RandomizedSearchCV
from imblearn.over_sampling import SMOTE
accuracy_scores = []
classification_reports = []
count_fold=1
for i, (train_index, val_index) in enumerate(sgkf.split(padded_sequences, y, X2_df)):
    print(f"\n----------Awal Fold {count_fold}:----------\n")

    X_train, X_val = padded_sequences[train_index], padded_sequences[val_index]
    y_train, y_val = y[train_index], y[val_index]
    #1
    custom_estimator = CustomEstimator()
    random_search = RandomizedSearchCV(estimator=custom_estimator, param_distributions=param_dist, n_iter=10, cv=3)
    random_search.fit(X_train, y_train, validation_data=(X_val, y_val), verbose=2)
    #2
    best_model = random_search.best_estimator_
    #3
    best_model_params = random_search.best_estimator_.get_params()
    print(f"\n{best_model_params}\n")
    #4
    from keras.callbacks import EarlyStopping,ReduceLROnPlateau
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, 
                                         restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, 
                                        patience=5, min_lr=0.00001)
    #5
    history = best_model.model.fit(X_train, y_train, epochs=1000,
                    validation_data=(X_val, y_val), callbacks=[early_stopping,reduce_lr])
    #6
    loss, accuracy = best_model.model.evaluate(X_val, y_val, verbose=0)
    print(f'loss: {loss:.2f}')
    print(f'Accuracy: {accuracy*100:.2f}%')
    accuracy_scores.append(accuracy)
    #7
    from sklearn.metrics import classification_report
    y_pred = best_model.model.predict(X_val)
    y_pred_classes = np.argmax(y_pred, axis=1)
    target_names =[str(cls) for cls in label_encoder.classes_]
    report=classification_report(y_val,y_pred_classes,target_names=target_names,output_dict=True,zero_division=1)
    print(classification_report(y_val,y_pred_classes,target_names=target_names,zero_division=1))
    classification_reports.append(report)

    print(f"----------Akhir fold {count_fold}----------\n\n")
    count_fold=count_fold+1
```

```python
print("Accuracy Scores:", accuracy_scores)
print("Mean Accuracy:", np.mean(accuracy_scores))
print("Standard Deviation of Accuracy:", np.std(accuracy_scores))
```
